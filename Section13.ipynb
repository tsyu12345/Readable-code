# 13章 短いコードを書く

> 最も読みやすいコードは、何も書かれていないコードだ

## 13.1 その機能の実装について悩まないで -きっと必要ないから-

新しいクラスの実装などで、追加するか悩んだような機能の実装をする場合、その機能追加で将来的に必要になるであろう保守や、ドキュメントの作成時間のコストを考えるべき。

ひょっとしたら、その機能が他のプログラマーの混乱を招くかもしれない。

## 13.2 質問と要求の分散

使う範囲が限定されているのであれば、本来実装しなければならない複雑な機能の実装が本当に必要かどうかを考えるべき。

## 13.3 コードを小さく保つ

コードの書く量（サイズ）が小さければ小さいほど、開発にかかる労力が少ないのは当然。
特に、プロジェクト初期の段階では、ファイルは１つや２つで済むので、バグの発見などは容易だ。

しかし、プロジェクトが大きくなるにつれて、ファイル数が増えていくと、バグの発見が難しくなる。
コードが分散し、依存関係も複雑になるからだ。

つまり、我々が注意することは、プロジェクトが大きくなってもいいようにコード量をできるだけ小さく維持することだ。



## 13.4 身近なライブラリに親しむ

ライブラリを使うことで、コードを短くすることができる。
特に、言語に標準で組み込まれているいわゆる標準ライブラリは、コードを書くうえでよく使われる機能を提供している事が多い。

例えば、配列のユニーク化などは、Pythonの場合、set()を使うことで簡単に実装できる。

```python
>>> a = [1, 2, 3, 4, 5, 5, 5, 6, 7, 8, 9, 9]
>>> set(a)
{1, 2, 3, 4, 5, 6, 7, 8, 9}
```
これを知らなければ、自身で`unique()`関数を実装する必要があったかもしれない。

そして、そのような手間は、コードの量を増やすことに繋がり、保守性も同時にそこなる。

また標準ライブラリは、長い時間をかけてきて保守、修正、ドキュメント化されているので、自身でそのような機能を実装するよりも圧倒的に良い。

## 13.6 まとめ

* 不必要な機能を除外する。過剰な機能はいらない。
* 最も簡単に問題を解決できる要求を考える
* 定期的にすべてのAPIを読んで、標準ライブラリに慣れ親しんでおこう。
